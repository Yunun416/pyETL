```python
# import urllib.request as req #用這個url連接中文會有問題
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd


keyword = input("keyword")
page = int(input("pages"))

data = []
for i in range(1, page+1):
    

    url = "https://www.104.com.tw/jobs/search/?ro=0&kwop=7&keyword=" + keyword +"&expansionType=area%2Cspec%2Ccom%2Cjob%2Cwf%2Cwktm&order=15&asc=0&page=" + str(i) + "&mode=s&jobsource=2018indexpoc&langFlag=0&langStatus=0&recommendJob=1&hotJob=1"
    headers={
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
    }


    result = requests.get(url, headers=headers).text
    soup = BeautifulSoup(result, 'html.parser')


    names = soup.find_all("article", class_="b-block--top-bord job-list-item b-clearfix js-job-item")
    # print(names)


    for name in names:
        job = name.find('a', class_="js-job-link").text
#         print(job)
        company = name.get("data-cust-name")
#         print(company)

        company_information = name.find("div", class_="job-list-tag b-content").text.strip()
        company_information_0 = company_information.split()
#         print(company_information_0)
        sal = company_information_0[0]
#         print(sal)

        locations = name.find("ul", class_="b-list-inline b-clearfix job-list-intro b-content")
        location = locations.li.string
#         print(location)
        more = ' , '.join(company_information_0[1 :])
#         print(more)
#         print("=========")
            
        job_data = {"職缺名稱": job, "公司名稱": company, "薪資": sal, "工作地點": location, "更多": more}
        data.append(job_data)
#     print("Next page")
print(data)
```

```python
columns = ["職缺名稱", "公司名稱", "薪資", "工作地點", "更多"]
df = pd.DataFrame(data = data, columns = columns)
df
```

```python
df.to_csv("104_job_.xls",encoding= "utf-8-sig", index=False)
```
